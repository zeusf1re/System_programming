## Математическое обоснование алгоритма

Для оптимизации вычислений и предотвращения переполнения разрядной сетки (при вычислении факториалов больших чисел) перейдем к рекурентному выражению n-го члена ряда через n-1-й.

Ращложение по Тейлору:
$(1 + 2x^2)e^{x^2}$:
$$ S(x) = 1 + 3x^2 + \dots + \frac{2n+1}{n!} x^{2n} + \dots = \sum_{n=0}^{\infty} a_n $$

n-ый член ряда:
$$ a_n = \frac{2n+1}{n!} x^{2n} $$

Предыдущий (n-1):
$$ a_{n-1} = \frac{2(n-1)+1}{(n-1)!} x^{2(n-1)} = \frac{2n-1}{(n-1)!} x^{2n-2} $$

Коэффициент перехода $K_n = \frac{a_n}{a_{n-1}}$:

$$ \frac{a_n}{a_{n-1}} = \frac{\frac{2n+1}{n!} x^{2n}}{\frac{2n-1}{(n-1)!} x^{2n-2}} $$

После группировки и упрощений:

$$ a_n = a_{n-1} \cdot \frac{2n+1}{2n-1} \cdot \frac{x^2}{n} $$

Учтем, что начинать надо с n = 1, при этом первый член равен 1, суммма ряда(для n = 1), соответсвтенно, тоже равна 1.
